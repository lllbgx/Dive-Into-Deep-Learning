```python
import torch
from torch import nn

class LeNet(nn.Module):
    def __init__(self):
        super(LeNet, self).__init__()
        
        # ==========================================
        # 第一部分：特征提取器 (卷积层 + 池化层)
        # ==========================================
        # 假设输入图片 X 的形状是:[batch_size, 1, 28, 28] (1个通道，高28，宽28)
        
        # TODO 1: 定义第一个卷积层
        # 要求：输入通道为 1，输出通道为 6，卷积核大小为 5，四周填充(padding)为 2
        # 提示：经过这层后，由于 padding=2，图像大小变为 28 - 5 + 2*2 + 1 = 28
        # 形状变化: [B, 1, 28, 28] ->[B, 6, 28, 28]
        self.conv1 = nn.Conv2d(1,6,5,padding=2)
        
        # TODO 2: 定义第一个最大池化层
        # 要求：核大小为 2，步幅(stride)为 2
        # 提示：图像宽高减半。形状变化:[B, 6, 28, 28] -> [B, 6, 14, 14]
        self.pool1 = nn.AvgPool2d(2)
        
        # TODO 3: 定义第二个卷积层
        # 要求：输入通道为 6，输出通道为 16，卷积核大小为 5，不填充(padding默认为0)
        # 提示：自己算一下图像大小变成多少了？公式: (输入大小 - 卷积核大小 + 2*padding) / 步幅 + 1
        # 形状变化:[B, 6, 14, 14] -> [B, 16, ?, ?] 
        self.conv2 = nn.Conv2d(6,16,5)
        
        # TODO 4: 定义第二个最大池化层
        # 要求：核大小为 2，步幅(stride)为 2
        # 形状变化: [B, 16, ?, ?] -> [B, 16, ?, ?]
        self.pool2 = nn.AvgPool2d(2)
        
        # ==========================================
        # 第二部分：分类器 (全连接层)
        # ==========================================
        self.relu = nn.ReLU()
        # TODO 5: 定义一个展平层（用于将 4D 张量变成 2D 张量，喂给全连接层）
        self.flatten = nn.Flatten()
        
        # TODO 6: 定义第一个全连接层 (极其关键！)
        # 思考：经过上面的所有层，图像最后变成了 16 个通道，每个通道的宽高是多少？
        # 这里的 in_features 必须等于 16 * 高 * 宽！算对这个，你就过关了！
        # 输出特征数设为 120。
        self.fc1 = nn.Linear(in_features=16*5*5, out_features=120)
        
        # TODO 7: 定义第二、第三个全连接层
        # 要求：fc2 输入 120，输出 84；fc3 输入 84，输出 10 (10分类)
        self.fc2 = nn.Linear(120,84)
        self.fc3 = nn.Linear(84,10)

    def forward(self, x):
        # ==========================================
        # 第三部分：前向传播拼装
        # ==========================================
        # 提示：按顺序通过 conv1 -> relu -> pool1 -> conv2 -> relu -> pool2
        # 然后 flatten 展平 -> fc1 -> relu -> fc2 -> relu -> fc3
        
        # TODO 8: 按照网络设计，手写出完整的前向传播流程，并返回最终结果
        x = self.conv1(x)
        x = self.relu(x)
        x = self.pool1(x)
        # 接着写...
        x = self.conv2(x)
        x = self.relu(x)
        x = self.pool2(x)
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.fc3(x)
        return x

# ==========================================
# 终极测试：形状驱动开发 (SDD)
# ==========================================
if __name__ == "__main__":
    # 实例化模型
    net = LeNet()
    
    # 1. 造一个假的图像输入：batch_size=1, 通道=1, 高=28, 宽=28
    X = torch.rand(size=(1, 1, 28, 28), dtype=torch.float32)
    
    # 2. 扔进网络，打印最终输出形状
    out = net(X)
    
    # 如果你的代码全对，这里绝对不会报错，而且会打印出 [1, 10]
    print(f"输入形状: {X.shape}")
    print(f"输出形状: {out.shape}")
```
